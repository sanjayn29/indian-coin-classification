```python
# ==========================================
# 1. Mount Google Drive and Import Libraries
# ==========================================
from google.colab import drive
import os
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
import matplotlib.pyplot as plt

# Mount Google Drive to access your dataset
drive.mount('/content/drive')

# Define the base directory path matching your dataset structure
DATA_DIR = '/content/drive/MyDrive/Dataset-net'

# Verify the path and classes exist
if os.path.exists(DATA_DIR):
    print("Dataset directory found!")
    print("Classes detected:", os.listdir(DATA_DIR))
else:
    print(f"Error: Directory {DATA_DIR} not found. Please check your folder structure.")

# ==========================================
# 2. Hyperparameters & Configuration
# ==========================================
# EfficientNetB0 natively expects 224x224 input images
IMG_SIZE = (224, 224)
BATCH_SIZE = 16  # Small batch size due to a small dataset (total ~200 images)
EPOCHS = 20
NUM_CLASSES = 4

# ==========================================
# 3. Load Dataset & Train-Validation Split
# ==========================================
# image_dataset_from_directory automatically infers labels based on folder names
print("\n--- Loading Training Dataset ---")
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,  # 80% for training, 20% for validation (~40 images)
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='categorical'  # Suitable for multi-class classification
)

print("\n--- Loading Validation Dataset ---")
val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='categorical'
)

# Extract class names to verify order mapping
class_names = train_ds.class_names
print("\nClass mapping verified:", {i: name for i, name in enumerate(class_names)})

# Optimize data loading pipeline performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

# ==========================================
# 4. Data Augmentation Pipeline
# ==========================================
# CRITICAL: Coins can be rotated or flipped in any direction.
# Augmentation creates synthetic variations to overcome the 50-image limit.
data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(factor=0.5),  # Full circle rotations
    layers.RandomZoom(height_factor=0.2),
    layers.RandomTranslation(height_factor=0.1, width_factor=0.1),
    layers.RandomContrast(factor=0.2),
])

# ==========================================
# 5. Build EfficientNetB0 Model Architecture
# ==========================================
# EfficientNet has internal rescaling, so we do not manually scale images to [0,1]
inputs = layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

# Apply the data augmentation to input images
x = data_augmentation(inputs)

# Load EfficientNetB0 pre-trained on ImageNet, excluding its top dense classification layers
base_model = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=x)

# Freeze the pre-trained weights to preserve learned feature extractors
base_model.trainable = False

# Rebuild the top custom classification layers for Indian Coins
x = layers.GlobalAveragePooling2D()(base_model.output)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)  # Dropout layer to mitigate overfitting risks

# Final Softmax layer outputting probabilities for 4 classes
outputs = layers.Dense(NUM_CLASSES, activation='softmax', name="pred")(x)

model = models.Model(inputs, outputs)

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==========================================
# 6. Model Training
# ==========================================
print("\n--- Starting Training ---")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)

# ==========================================
# 7. Evaluate and Plot Accuracy/Loss Curves
# ==========================================
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()

# ==========================================
# 8. Save the Trained Model
# ==========================================
# Saves the trained weights onto your Drive folder
model_save_path = os.path.join(DATA_DIR, 'indian_coin_efficientnet.h5')
model.save(model_save_path)
print(f"\nModel successfully saved to: {model_save_path}")
```

### Visualize Sample Predictions

Let's visualize some predictions on the validation set to see how the model performs.

```python
import numpy as np

# Get a batch of validation data
for images, labels in val_ds.take(1):
    break

# Make predictions
predictions = model.predict(images)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(labels.numpy(), axis=1)

plt.figure(figsize=(15, 10))
for i in range(BATCH_SIZE):
    plt.subplot(4, 4, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    predicted_label = class_names[predicted_classes[i]]
    true_label = class_names[true_classes[i]]
    color = "green" if predicted_label == true_label else "red"
    plt.title(f"True: {true_label}\nPred: {predicted_label}", color=color)
    plt.axis("off")
plt.tight_layout()
plt.show()
```

### Fine-tuning the Model

To further improve accuracy, we can unfreeze the top layers of the pre-trained EfficientNetB0 model and continue training with a much smaller learning rate.

```python
# Unfreeze the base model
base_model.trainable = True

# Recompile the model with a much smaller learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),  # Significantly smaller learning rate
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

print("\n--- Starting Fine-tuning Training ---")
history_fine_tune = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10  # Train for a few more epochs
)
```

### Evaluate Fine-tuned Model and Plot Curves

Let's evaluate the performance after fine-tuning.

```python
# Combine histories
total_acc = history.history['accuracy'] + history_fine_tune.history['accuracy']
total_val_acc = history.history['val_accuracy'] + history_fine_tune.history['val_accuracy']

total_loss = history.history['loss'] + history_fine_tune.history['loss']
total_val_loss = history.history['val_loss'] + history_fine_tune.history['val_loss']

epochs_range_fine_tune = range(len(total_acc))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range_fine_tune, total_acc, label='Training Accuracy')
plt.plot(epochs_range_fine_tune, total_val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (Fine-tuned)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range_fine_tune, total_loss, label='Training Loss')
plt.plot(epochs_range_fine_tune, total_val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (Fine-tuned)')
plt.show()
```

```python
acc_fine_tune = history_fine_tune.history['accuracy']
val_acc_fine_tune = history_fine_tune.history['val_accuracy']
loss_fine_tune = history_fine_tune.history['loss']
val_loss_fine_tune = history_fine_tune.history['val_loss']

epochs_range_fine_tune_only = range(len(acc_fine_tune))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range_fine_tune_only, acc_fine_tune, label='Training Accuracy')
plt.plot(epochs_range_fine_tune_only, val_acc_fine_tune, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (Fine-tuning Phase Only)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range_fine_tune_only, loss_fine_tune, label='Training Loss')
plt.plot(epochs_range_fine_tune_only, val_loss_fine_tune, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (Fine-tuning Phase Only)')
plt.show()
```

```python
# Saves the fine-tuned model
model_save_path_fine_tune = os.path.join(DATA_DIR, 'indian_coin_efficientnet_fine_tuned.h5')
model.save(model_save_path_fine_tune)
print(f"\nFine-tuned model successfully saved to: {model_save_path_fine_tune}")
```
